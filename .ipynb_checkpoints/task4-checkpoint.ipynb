```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
items_df = pd.read_csv("data/olist_order_items_dataset.csv")
orders_df = pd.read_csv("data/olist_orders_dataset.csv")

products_df = pd.read_csv("data/olist_products_dataset.csv")
tans_df= pd.read_csv("data/product_category_name_translation.csv")
reviews_df = pd.read_csv("data/olist_order_reviews_dataset.csv")
```

```python
df = products_df.merge(tans_df,on="product_category_name",how="left")
df.head()
```

```python
df = df.merge( items_df,on="product_id",how="left")
df.head()
```

```python
df = df.merge(orders_df,on="order_id",how="left")
df.head()
```

```python
df = df.merge( reviews_df,on="order_id",how="left")
df.head()
```

```python
df.columns
```

```python
final_df = df[
    [
        "product_id",
        "product_category_name",
        "product_category_name_english",
        "review_score",
        "review_comment_title",
        "review_comment_message"
    ]
]

print(final_df.head())
print(final_df.shape)
```

```python
df = final_df.drop_duplicates().dropna()
df.head()
```

```python
df.drop(columns=['product_id','product_category_name'],inplace=True)
```

```python
df.drop(columns=['review_score'],inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop_duplicates()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df['product_category_name_english'].unique()
```

```python
df.to_csv("data/clean_olist_dataset.csv", index=False)
```
